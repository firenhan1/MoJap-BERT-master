```python
from glob import glob
from tqdm import tqdm
import shutil
import codecs
import os

BLOCKSIZE = 1048576
input_file = 'MAI98B.txt'
output_file = 'splitted data/'
```

```python
for sourceFileName in tqdm(glob(input_file, recursive=True)):
    baseName = os.path.basename(sourceFileName)
    with codecs.open(sourceFileName, "r", "shift-jisx0213", errors='ignore') as sourceFile:
        with codecs.open(output_file + baseName, "w", "utf-8") as targetFile:
            while True:
                contents = sourceFile.read()
                if not contents:
                    break
                targetFile.write(contents)
                
```
